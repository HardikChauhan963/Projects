##Objective:

"This Python program empirically demonstrates the Central Limit Theorem (CLT) by analyzing the distribution of sample means derived from various probability distributions, including binomial, Poisson, uniform, and Pareto. Through repeated sampling and visualization, the code showcases the convergence of sample mean distributions towards a normal distribution, regardless of the underlying population distribution, validating the fundamental principles of the CLT."

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def population_distribution():

  #Generating 1000 samples with each having 10 trials with probability of success being 0.2
  binomial = np.random.binomial(n=50, p=0.2, size=10000)

  #Generating 1000 samples with having range between [3,5]
  uniform = np.random.uniform(3, 5, 10000)

  #Generates 1000 random values from a Poisson distribution with an average event rate of 4 per interval.
  poisson = np.random.poisson(lam=5, size=10000)

  #Generates Pareto's distribution i.e. 60-40 distribution
  pareto = random.pareto(a=4, size=10000)


  sample_binomial = []
  sample_uniform = []
  sample_poisson = []
  sample_pareto = []

  #Sampling
  for i in range (1000):
    size = 100
    sample_binomial.append(np.mean(np.random.choice(binomial, size, replace= False)))
    sample_uniform.append(np.mean(np.random.choice(uniform, size, replace=False)))
    sample_poisson.append(np.mean(np.random.choice(poisson, size, replace= False)))
    sample_pareto.append(np.mean(np.random.choice(pareto, size, replace= False)))



  #Plotting
  fig, ax = plt.subplots(4, 2, figsize=(10, 10))
  distributions = [
  (binomial, "Binomial Distribution"),
  (sample_binomial, "Sampling Distribution from Binomial"),
  (uniform, "Uniform Distribution"),
    (sample_uniform, "Sampling Distribution from Uniform"),
    (poisson, "Poisson Distribution"),
    (sample_poisson, "Sampling Distribution from Poisson"),
    (pareto, "Pareto Distribution"),
    (sample_pareto, "Sampling Distribution from Pareto"),
  ]

  for i, (data, title) in enumerate(distributions):
    row, col = divmod(i, 2)
    sns.histplot(data, bins=30, kde=True, ax=ax[row, col])
    ax[row, col].set_title(title)
    ax[row, col].text(
        0.8, 0.8, f"Mean: {round(np.mean(data), 2)}",
        fontsize=12, ha='center', va='center', transform=ax[row, col].transAxes
    )
    ax[row, col].text(
        0.8, 0.7, f"Variance: {round(np.var(data) * (100 if 'Sampling' in title else 1), 2)}",
        fontsize=12, ha='center', va='center', transform=ax[row, col].transAxes
    )

  plt.tight_layout()
  plt.show()



```

```python
population_distribution()
```
