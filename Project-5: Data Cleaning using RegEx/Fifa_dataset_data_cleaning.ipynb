```python
#STEP-1: Extracting Club name from 'Team & Contract' Column

df1['Club'] = df1['Team & Contract'].str.extract(r'\n(.+?)\n', expand=False).str.strip()
```

```python
#Extracting 'Contract_start' date from 'Team & Contract' column
#The regex \d{4} is used to match exactly four digits in a string.

df1['Contract_start'] = df1['Team & Contract'].str.extract(r'(\d{4})')
```

```python
#Extracting 'Contract_end' date from 'Team & Contract' column
#The regex (\d+)\s*$ is used to match the last number at the end of a string, possibly followed by trailing whitespace.

df1['Contract_end'] = df1['Team & Contract']. str.extract((r'(\d+)\s*$'), expand = False).str.strip()
```

```python
#Converting "Contract_start" and "Contract_end" dates to type int

df1["Contract_start"] = pd.to_numeric(df1["Contract_start"], errors="coerce").fillna(0).astype(int)
df1["Contract_end"] = pd.to_numeric(df1["Contract_end"], errors="coerce").fillna(0).astype(int)
```

```python
#STEP-2: Renaming columns in df1 to match df2
df1.rename(columns = {'foot' : 'Preferred Foot', 'BP' : 'Best Position'}, inplace = True)
```

```python
#Cleaning Club column in df2
df2['Club'] = df2['Club'].str.strip()
```

```python
#Splitting the Contract column to Contract_start and Contract_end columns
df2['Contract_start'] = df2['Contract'].str.extract(r'(\d{4})', expand = False).str.strip()
df2['Contract_end'] = df2['Contract'].str.extract(r'(\d{4})\s*$', expand = False).str.strip()
```

```python
#STEP-3: Now we will join df2 below df1

df1 = pd.concat([df1, df2], axis = 0, ignore_index = True)
```

```python
#Displaying all columns.

with pd.option_context('display.max_rows', None):
  print(df1.isnull().sum())
```

```python
#Dropping all columns in which NAN values are greater then 50% in their respective columns.

for i in df1.columns:
  if df1[i].isnull().sum() >= 0.5*len(df1[i]):
    df1.drop(i, axis = 1, inplace = True)
```

```python
#STEP-4: Dropping duplicates

df1.drop_duplicates(inplace = True)
```

```python
#Height is mixed (Some are in ft-inches and some values are in cms)
for index, i in df1['Height'].items(): # Iterate using index and value
    if isinstance(i, str) and "'" in i: # Check if value is string and contains "'"
        feet, inches = i.split("'")
        inches = inches.replace('"', '')
        inches = int(inches)
        feet = int(feet)
        cms = round((feet * 30.48) + (inches * 2.54))
        df1.loc[index, 'Height'] = cms # Update 'Height' using index

```

```python
# Removing "cm" string from the remaining height values

for i in df1['Height']:
  if isinstance(i, str) and 'cm' in i:
    df1['Height'] = df1['Height'].str.replace('cm', '')
```

```python
#Also renaming the column from Height to 'Height_cms'

df1.rename(columns = {'Height' : 'Height_cms'}, inplace = True)
```

```python
#Cleaning weight columns.
#Weight column has mixed values (some are in 'lbs' and some are in 'kg')


for index, i in df1['Weight'].items():
  if isinstance(i,str):
    if 'lbs' in i:
        weight_lbs = i.replace('lbs', '').strip()
        weight_lbs = int(weight_lbs)
        weight_kg = round(weight_lbs* 0.453592, 2)
        df1.loc[index, 'Weight'] = weight_kg
    elif 'kg' in i:
        weight_kg = i.replace('kg', '').strip()
        weight_kg= int(weight_kg)
        df1.loc[index, 'Weight'] = weight_kg
```

```python
#Also renaming the column from Weight to 'Weight_kgs'

df1.rename(columns = {'Weight' : 'Weight_kgs'}, inplace = True)
```

```python
#Function to clean values in "Value", "Wage" and "Release Clause" columns and converting them in milions

def convert_currency(x):
  for index, i in df1[x].items():
    if isinstance(i,str):
      i = i.replace('€', '').strip()

      if 'K' in i:
        i = i.replace('K', '')
        i = float(i)
        i = i/1000

      elif 'M' in i:
        i = i.replace('M', '')
        i = float(i)

      elif 'K' not in i and 'M' not in i:
        i = float(i)

      df1.loc[index, x] = i


```

```python
convert_currency('Value')
convert_currency('Wage')
convert_currency('Release Clause')
```

```python
#Also renaming "Value", "Wage" and "Release Clause" to "Value_in_millions", "Wage_in_millions" and "Release Clause_in_millions"

df1.rename(columns = {'Value': "Value_in_millions",'Wage': "Wage_in_millions", 'Release Clause': "Release_Clause_in_millions" }, inplace = True)
```

```python
#Extracting the numeric value of "W/F" and "IR" columns and discarding the *
def converter(x):
    df1[x] = pd.to_numeric(df1[x].astype(str).str.strip().str.split().str[0], errors = 'coerce')
```
